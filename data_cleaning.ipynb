# Cleaned data

```python
import os
import shutil
import pandas as pd

SOURCE_DIRS = [
    "stock_market_data/forbes2000/csv",
    "stock_market_data/nasdaq/csv",
    "stock_market_data/nyse/csv",
    "stock_market_data/sp500/csv"
]

DEST_DIR = "backend/data"
os.makedirs(DEST_DIR, exist_ok=True)

THRESHOLD = 1_000_000

for ticker in all_files:
    copied = False
    filename = f"{ticker}.csv"

    for src_dir in SOURCE_DIRS:
        src_path = os.path.join(src_dir, filename)

        if os.path.exists(src_path):
            try:
                df = pd.read_csv(src_path)
            except Exception:
                print(f"ERROR reading {filename}")
                break

            # Convert all columns to numeric where possible
            numeric_df = df.apply(pd.to_numeric, errors="coerce")

            # Keep rows where ALL numeric values are within range
            mask = (numeric_df <= THRESHOLD) & (numeric_df >= -THRESHOLD)
            valid_rows = mask.all(axis=1)

            cleaned_df = df[valid_rows]

            # Save cleaned file
            dst_path = os.path.join(DEST_DIR, filename)
            cleaned_df.to_csv(dst_path, index=False)

            copied = True
            break

    if not copied:
        print(f"WARNING: {filename} not found in any source folder")
```

```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt

data_dir = Path("./backend/data")

all_values = []

for file in data_dir.glob("*.csv"):
    try:
        df = pd.read_csv(file)
    except Exception:
        continue

    # Detect column name
    col_name = None
    for candidate in ["Adjusted Close"]:
        if candidate in df.columns:
            col_name = candidate
            break

    if col_name is None:
        continue

    # Convert to numeric
    df[col_name] = pd.to_numeric(df[col_name], errors="coerce")
    df = df.dropna(subset=[col_name])


    all_values.extend(df[col_name].values)
plt.figure()
plt.boxplot(all_values)
plt.title("Boxplot of Adjusted Close (All Stocks)")
plt.ylabel("Adjusted Close")
plt.show()
```
