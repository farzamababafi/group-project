# Cleaned data

```python
import os
import pandas as pd

SOURCE_DIRS = [
    "stock_market_data/forbes2000/csv",
    "stock_market_data/nasdaq/csv",
    "stock_market_data/nyse/csv",
    "stock_market_data/sp500/csv"
]

DEST_DIR = "backend/data"
os.makedirs(DEST_DIR, exist_ok=True)

THRESHOLD = 1000000
numeric_cols = ["Open", "High", "Low", "Close", "Adjusted Close"]

dropped_files = []
kept_files = []

for ticker in all_files:
    filename = f"{ticker}.csv"
    found = False

    for src_dir in SOURCE_DIRS:
        src_path = os.path.join(src_dir, filename)

        if not os.path.exists(src_path):
            continue

        found = True

        try:
            df = pd.read_csv(src_path)
        except Exception:
            print(f"ERROR reading {filename}")
            dropped_files.append(filename)
            break

        # Ensure required columns exist
        existing_cols = [col for col in numeric_cols if col in df.columns]
        if len(existing_cols) == 0:
            print(f"❌ Dropped {filename} (no valid columns)")
            dropped_files.append(filename)
            break


        # Convert to numeric
        df[existing_cols] = df[existing_cols].apply(pd.to_numeric, errors="coerce")

        # 🚨 THIS IS THE KEY CHECK
        invalid_mask = (
            (df[existing_cols] > THRESHOLD) |
            (df[existing_cols] < -THRESHOLD) |
            (df[existing_cols].isna())   # also treat NaN as invalid
        )

        if invalid_mask.any().any():
            print(f"❌ Dropped {filename} (invalid values found)")
            
            # Optional: show one bad row for debugging
            bad_row = df[invalid_mask.any(axis=1)].iloc[0]
            print(bad_row[existing_cols], "\n")

            dropped_files.append(filename)
            break

        # ✅ File is clean → save it
        dst_path = os.path.join(DEST_DIR, filename)
        df.to_csv(dst_path, index=False)
        kept_files.append(filename)
        break

    if not found:
        print(f"WARNING: {filename} not found")

print("\n---- SUMMARY ----")
print(f"Kept files: {len(kept_files)}")
print(f"Dropped files: {len(dropped_files)}")
```

```python
import pandas as pd
from pathlib import Path

data_dir = Path("./backend/data/")

files = sorted(
    p for p in data_dir.iterdir()
    if p.is_file() and p.suffix == ".csv"
)

THRESHOLD = 8_000_000_000  # 8 billion

results = []

for file in files:
    try:
        df = pd.read_csv(file)
    except Exception:
        continue

    # Detect column name
    col_name = None
    for candidate in ["Adjusted Close", "Adj Close"]:
        if candidate in df.columns:
            col_name = candidate
            break

    if col_name is None:
        continue

    # Clean numeric
    df[col_name] = pd.to_numeric(df[col_name], errors="coerce")
    df = df.dropna(subset=[col_name])

    if df.empty:
        continue

    # Min / Max
    min_value = df[col_name].min()
    max_value = df[col_name].max()

    min_row = df.loc[df[col_name].idxmin()]
    max_row = df.loc[df[col_name].idxmax()]

    has_negative = min_value < 0
    has_huge = max_value > THRESHOLD

    results.append({
        "file": file.name,
        "min": min_value,
        "max": max_value,
        "has_negative": has_negative,
        "has_huge": has_huge,
    })

    # Print anomalies
    if has_negative:
        print(f"❌ NEGATIVE in {file.name}")
        print(min_row[["Date", col_name]], "\n")

    if has_huge:
        print(f"🚨 HUGE VALUE in {file.name}")
        print(max_row[["Date", col_name]], "\n")

print("---- SUMMARY ----")
print(f"Files checked: {len(results)}")
print(f"Negative count: {sum(r['has_negative'] for r in results)}")
print(f"Huge value count: {sum(r['has_huge'] for r in results)}")
```
