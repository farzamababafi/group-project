```python
from pathlib import Path
import os
import shutil
```

# Getting the name of stocks in each folder

```python
forbes2000_files = sorted(
    p.stem
    for p in Path("./stock_market_data/forbes2000/csv/").iterdir()
    if p.is_file() and p.suffix == ".csv")
print(forbes2000_files) 
```

```python
nasdaq_files = sorted(
    p.stem
    for p in Path("./stock_market_data/nasdaq/csv/").iterdir()
    if p.is_file() and p.suffix == ".csv"
)

print(nasdaq_files)
```

```python
nyse_files = sorted(
    p.stem
    for p in Path("./stock_market_data/nyse/csv/").iterdir()
    if p.is_file() and p.suffix == ".csv"
)

print(nyse_files)
```

```python
sp500_files = sorted(
    p.stem
    for p in Path("./stock_market_data/sp500/csv/").iterdir()
    if p.is_file() and p.suffix == ".csv"
)

print(sp500_files)
```

# Distinct stocks

```python
all_files = sorted(set(
    forbes2000_files
    + nasdaq_files
    + nyse_files
    + sp500_files
))
print(len(all_files))
```

# Cleaned data

```python
SOURCE_DIRS = [
    "stock_market_data/forbes2000/csv",
    "stock_market_data/nasdaq/csv",
    "stock_market_data/nyse/csv",
    "stock_market_data/sp500/csv"
]

DEST_DIR = "backend/data"
os.makedirs(DEST_DIR, exist_ok=True)
for ticker in all_files:
    copied = False
    filename = f"{ticker}.csv"

    for src_dir in SOURCE_DIRS:
        src_path = os.path.join(src_dir, filename)

        if os.path.exists(src_path):
            dst_path = os.path.join(DEST_DIR, filename)
            shutil.copy2(src_path, dst_path)
            copied = True
            break

    if not copied:
        print(f"WARNING: {filename} not found in any source folder")
```
